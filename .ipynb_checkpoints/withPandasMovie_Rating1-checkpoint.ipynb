```python
# creating a sub datafram from orginal which contains title,industry,language as its columns
df_sub=movies_df[["title","industry","language"]]
df_sub
```

```python
# printing data of all movies that were released from 2000 to 2010 excluding 2010
movies_df[(movies_df["release_year"]>=2000) & (movies_df["release_year"]<2010)]
```

```python
# printing the records that belong to Marvel Studios

only_marvel_df=movies_df[movies_df["studio"] == 'Marvel Studios']

only_marvel_df
```

```python
movies_df.info()
```

```python
# creating a dataframe and storing details of movies with maximum and minimum ratings only
# movies_df["imdb_rating"].max()->returns the max value of imdb
# movies_df["imdb_rating"].max()->returns the min value of imdb

max_min_ratedMovie_data=movies_df[(movies_df["imdb_rating"]==movies_df["imdb_rating"].max()) | (movies_df["imdb_rating"]==movies_df["imdb_rating"].min())]

max_min_ratedMovie_data
```
